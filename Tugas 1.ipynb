```
# 1. Install Library Pytrends
!pip install pytrends

# 2. Import Libraries yang Diperlukan
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt

# 3. Mengatur Koneksi ke Google Trends
pytrends = TrendReq(hl='id-ID', tz=360)

# 4. Kata Kunci yang Ingin Dicari
keywords = ["billiard"]

# 5. Mengatur Wilayah dan Periode Waktu
pytrends.build_payload(keywords, geo='ID-JB', timeframe='2024-01-01 2024-12-31')

# 6. Mendapatkan Data Minat dari Google Trends
data = pytrends.interest_over_time()

# 7. Membersihkan Data Jika Ada Kolom yang Tidak Diperlukan
data = data.drop(columns=['isPartial'])

# 8. Menampilkan Data Awal
print(data.head())

# 9. Visualisasi Data
plt.figure(figsize=(10,6))
plt.plot(data.index, data['billiard'], marker='o', color='b', label='Minat terhadap Billiard')
plt.xlabel('Tanggal')
plt.ylabel('Tingkat Minat')
plt.title('Tren Minat Olahraga Billiard di Jabodetabek (2024)')
plt.legend()
plt.grid(True)
plt.show()

# 10. Menyimpan Data ke CSV (Opsional)
data.to_csv("minat_billiard_jabodetabek_2024.csv")
```
